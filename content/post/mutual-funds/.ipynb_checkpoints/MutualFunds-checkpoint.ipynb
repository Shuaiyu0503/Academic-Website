```python
crsp_name = crsp[['CRSP_FUNDNO','FUND_NAME','FIRST_OFFER_DT']].dropna().drop_duplicates().reset_index(drop = True)
crsp_name.tail()
crsp_name['MATCH_NAME'] = crsp_name['FUND_NAME'].str.lower().str.split().str[0]
crsp_name['INCEPTION'] = crsp_name['FIRST_OFFER_DT'] + MonthEnd(0)
crsp_name = crsp_name[['CRSP_FUNDNO','INCEPTION','MATCH_NAME','FUND_NAME']]
crsp_name[['FUND_NAME','MATCH_NAME']].nunique()
```

```python
from fuzzywuzzy import fuzz
name_match['SIM'] = name_match.apply(lambda row: fuzz.token_set_ratio(row['FUND STANDARD NAME'], row['FUND_NAME']), axis = 1)
name_match = name_match.sort_values(['CRSP_FUNDNO','SIM','SECID']).reset_index(drop = True)
name_match.head(50) 
name_match.tail(50) 
```

```python
match_res1 = name_match[['CRSP_FUNDNO','SECID','SIM']]
match_res1['RANK1'] = match_res1.groupby(['CRSP_FUNDNO']).SIM.rank(ascending = False)
match_res2 = name_match[['CRSP_FUNDNO','SECID','SIM']]
match_res2['RANK2'] = match_res2.groupby(['SECID']).SIM.rank(ascending = False)
match_res = pd.merge(match_res1[match_res1['RANK1'] <= 5], match_res2[match_res2['RANK2'] <= 5].drop('SIM', axis = 1), 
                     on = ['CRSP_FUNDNO','SECID'], how = 'inner')  
```

```python
match_res = match_res.sort_values(['CRSP_FUNDNO','SIM']).drop_duplicates(['SECID'], keep = 'last')        
match_res[['CRSP_FUNDNO','SECID']].nunique()
```

```python
ret_tna_cusip = pd.merge(crsp_ret_tna, crsp_ms[['CRSP_FUNDNO','SECID_CUSIP']].dropna().drop_duplicates(),
                         on = 'CRSP_FUNDNO', how = 'inner')
ret_tna_cusip = pd.merge(ret_tna_cusip, ms_ret_tna.rename(columns = {'SECID':'SECID_CUSIP'}), 
                         on = ['SECID_CUSIP','YYYYMM'], how = 'inner')
ret_tna_cusip = ret_tna_cusip.drop_duplicates(['CRSP_FUNDNO','SECID_CUSIP','YYYYMM']).reset_index(drop = True)
ret_tna_cusip['Diff_Ret'] = (ret_tna_cusip['MRET_CRSP']-ret_tna_cusip['MRET_MS']).abs()
ret_tna_cusip['Diff_Tna'] = (ret_tna_cusip['MTNA_CRSP']-ret_tna_cusip['MTNA_MS']).abs()
ret_tna_cusip.tail()
ret_tna_cusip[['CRSP_FUNDNO','SECID_CUSIP']].nunique()
```

```python
screen1 = (ret_tna_cusip.groupby('CRSP_FUNDNO').Diff_Ret.quantile(.5) <= 0.001).to_frame(name = 'SCREEN1')
screen2 = (ret_tna_cusip.groupby('CRSP_FUNDNO').Diff_Tna.quantile(.5) <= 0.2).to_frame(name = 'SCREEN2')
screen = screen1.join(screen2).reset_index()
ret_tna_cusip = pd.merge(ret_tna_cusip, screen[screen['SCREEN1'] & screen['SCREEN2']]['CRSP_FUNDNO'], 
                         on = 'CRSP_FUNDNO', how = 'inner')
ret_tna_cusip.tail()
ret_tna_cusip[['CRSP_FUNDNO','SECID_CUSIP']].nunique()
```

```python
screen1 = (ret_tna_ticker.groupby('CRSP_FUNDNO').Diff_Ret.quantile(.5) <= 0.001).to_frame(name = 'SCREEN1')
screen2 = (ret_tna_ticker.groupby('CRSP_FUNDNO').Diff_Tna.quantile(.5) <= 0.2).to_frame(name = 'SCREEN2')
screen = screen1.join(screen2).reset_index()
ret_tna_ticker = pd.merge(ret_tna_ticker, screen[screen['SCREEN1'] & screen['SCREEN2']]['CRSP_FUNDNO'], 
                         on = 'CRSP_FUNDNO', how = 'inner')
ret_tna_ticker.tail()
ret_tna_ticker[['CRSP_FUNDNO','SECID_TICKER']].nunique()
```
